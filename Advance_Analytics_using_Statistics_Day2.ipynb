```python
import numpy as np
import pandas as pd
```

```python
# Set
a1 = np.array([3,7,6,1,8,9])
a2 = np.array([2,4,5,8,10,3,6])
a1 , a2
```

```python
# | (+) -> union1d() -- either in set one or set 2
np.union1d(a1,a2)
```

```python
# & (*) -> intersect1d() -- common to set 1 and set 2
np.intersect1d(a1,a2)
```

```python
#np.in1d(a1,a2)    #/tmp/ipykernel_10826/1595240845.py:1: DeprecationWarning: `in1d` is deprecated. Use `np.isin` instead.np.in1d(a1,a2)
np.isin(a1,a2)
```

```python
# setdiff1d()
np.setdiff1d(a1,a2)
```

```python
np.setdiff1d(a2,a1)
```

```python
# setxor1d() - union of unique element      # symmetric_difference -> {3, 7, 6, 1, 8, 9}.symmetric_difference({2,  4,  5,  8, 10,  3,  6}) = {1, 2, 4, 5, 7, 9, 10}
np.setxor1d(a1,a2)
```

```python
# Vectors
x1 = np.array([2,3,-4])
x2 = np.array([3,-2,1])
x1,x2
```

```python
# dot() -> dot product of 2 vectors
# Note : when dot product of 2 vectors is 0 then 2 vecotrs are perpendicular to each other
np.dot(x1,x2)
```

```python
# determinant -> linalg.det()
from numpy import linalg

x3 = np.array([[1,2,3],[0,1,2],[1,1,0]])
np.linalg.det(x3)
```

```python
# cross() -> cross product of 2 vecctors
np.cross(x1,x2)
```

# Series

```python
import pandas as pd
```

```python
# series creation
x1 = [4,8,4,3,2]
x1 = pd.Series(x1)
x1
```

```python
# labeling series
x1 = [4,8,4,3,2]
s1 = pd.Series(x1, index = ['I','II','III','IV','V'])
s1
```

```python
# index based extraction
s1.iloc[2]
```

```python
# label based extraction
s1.loc['III']
```

```python
print(s1)
print(s1.iloc[[1,3]])   # index based extraction of multiple values
print(s1.loc[['II','IV']])  # label based extraction of multiple values
```

```python
# labeling can contain duplicate labels
s2 = pd.Series([1,2,3,4,5],index=['A','A','B','A','B'])
print(s2,end="\n\n")

print(s2.iloc[3],end="\n\n")

print(s2.loc['A'],end="\n\n")

print(s2.loc[['A','B']])
```

```python
# merging or concatinating two series
print(s1,end="\n\n")

s3 = pd.Series([3,9],index=['VI',"VII"])
print(s3,end="\n\n")

s1 = pd.concat([s1,s3])
print(s1)
len(s1)
```

```python
print(s1.iloc[4])
s1.iloc[4]=9          # modify value based on index
print(s1,end="\n\n")

print(s1.loc['V'])
s1.loc['V']=11        # modify value based on label
print(s1)
```

```python
# modification on duplicate labels
print(s2,end='\n\n')
s2.loc['A']=4
print(s2,end='\n\n')
s2.iloc[1]=8
print(s2,end='\n\n')
```

```python
# deleting values from series with help of labels
s1 = s1.drop(['V'])         # droped vale at label V
print(s1,end='\n\n')
s1 = s1.drop(['VI','VII'])  # droped value with multiple label
print(s1,end='\n\n')
```

```python
# deleting values from series with help of index
print(s1,end='\n\n')
print(s1.drop(s1.index[2]))
print(s1,end='\n\n')
print(s1.drop(s1.index[[1,2]]))
print(s1,end='\n\n')
```

```python
s4 = pd.Series([2,3,2,2,4,4,3,4,4,4,5,5,2,4,5])
s5=s4.value_counts()    # counts the count of number and will display in descending order of values
print(s5,end='\n\n')
print(s5.loc[2],end='\n\n')
print(s5.iloc[2],end='\n\n')
```

```python
# if we dont know index and label just know values
print(s1)

print(np.where(s1==8))

print(s1.index[np.where(s1==8)[0]])

print(s1.index)

s1.iloc[np.where(s1==4)[0]]=1
s1.loc[s1.index[np.where(s1==8)[0]]]=10
s1.loc[s1.index[np.where(s1==3)]]=10
print(s1)
```

# DataFrame

```python
# DataFrame Creation

mydata = [['A',1234,20000],['B',5678,10000],['C',1357,8000],['D',4567,30000]]
print(len(mydata),end='\n\n')
print(len(mydata[0]),end='\n\n')

mydf = pd.DataFrame(mydata)
print(mydf,end='\n\n')

mydf = pd.DataFrame(mydata,columns=['Name','AccNo','Balance'])
print(mydf,end='\n\n')

temp = pd.DataFrame(mydata,columns=['Name','AccNo','Balance'],index=['Sujal','Chaitu','Omkie','Deva'])
print(temp,end='\n\n')
```

```python
print(mydf,end='\n\n')

print(mydf.iloc[[1,2]],end='\n\n')

print(mydf.loc[[1,2]],end='\n\n')

df1 = mydf.iloc[[1,2]]                        # it carry forwards the same label
print(df1,end='\n\n')
print(df1.iloc[1],end='\n\n')
print(df1.loc[1],end='\n\n')

df1 = mydf.iloc[[1,2]].reset_index(drop=True) # it will not carry forward the same label
print(df1,end='\n\n')

df1 = mydf.iloc[[1,2]].reset_index()  # if we dont mention drop=True then index column will be added
print(df1,end='\n\n')
```

```python
# DataFrame creation from dictionary
stu_df = {"Name":['A','B','C','D'],"RollNo":[1,2,3,4],"Marks":[50,60,75,64]}

df1 = pd.DataFrame(stu_df)
print(df1,end='\n\n')
```

```python
# adding rows into df
print(df1,end='\n\n')

newdf = pd.DataFrame([['E',7,80],['F',10,65]],columns=df1.columns)  # for adding we should have same column names.
print(newdf,end='\n\n')

df2 = pd.concat([df1,newdf]).reset_index(drop=True)
print(df2,end='\n\n')
```

```python
# adding new column to df
x1 = [67,89,75,62,59,82]
df2['Marks2']=x1
print(df2,end='\n\n')
```

```python
# to get shape of df
print(df2.shape)
print(df2.shape[0]) # no. of rows
print(df2.shape[1]) # no. of cols
```

```python
# accessing or modyfying perticular value use df.at[] -> at works same like loc i.e uses labels
df2.at[3,"Marks"]
df2.at[3,"Marks"] = 74
print(df2)
```

```python
# df.iat[] -> it is similar to iloc[]
print(df2.iat[1,3])
print(df2.iloc[1,3])
```

```python
# extracting or accesing complete row
print(df2.iloc[[0,3,5]],end='\n\n') # the labels were carry forwarded so use reset_index

df3 = df2.iloc[[0,3,5]].reset_index(drop=True)
print(df3,end='\n\n')
```

```python
# extracting or accesing complete column
mySubject = df2[['Name','Marks']]   # extract using column name
print(mySubject,end='\n\n')

print(df2.iloc[:,[1,2]],end='\n\n') # extract using column index

print(df2['Marks'])
print(df2.Marks)    # we can also extract single column like this if column name does not have space
```

```python
# removing rows from df
print(df2.drop([1],axis=0),end='\n\n')
print(df2.drop([2,4],axis=0) )
```

```python
# removing cols from df
print(df2.drop('Marks2',axis=1), end='\n\n' )
print(df2.drop(['RollNo','Marks'],axis=1), end='\n\n' )
```
