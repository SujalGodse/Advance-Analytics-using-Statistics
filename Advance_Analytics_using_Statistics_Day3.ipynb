Data Cleaning

```python
df = pd.read_excel("CDAC_DataBook.xlsx",sheet_name="EmpInfo")
df
```

```python
# find missing values
ind = np.where(df.Deptt.isnull())
print(ind)

df2 = df.iloc[ind[0]].reset_index(drop=True)      # select rows having missing values
print(df2,end='\n\n')

df1 = df.drop(ind[0]).reset_index(drop=True)      # select rows which are not having missing values
print(df1)
```

```python
print(df,end='\n\n')
ind = np.where((df.EmpID.isnull()) & (df.Deptt.isnull()))
print(df.iloc[ind])
```

```python
# Handlinf missing values
# 1) remove rows -> no. of rows are few
# 2) replace the missing values with some other values
# -> int - median value
# -> string - mode value
ind = np.where(df.Deptt.isnull())
print(df,end='\n\n')
df.loc[ind[0],'Deptt']='Fresher'   # replace the missing value with word 'Fresher'
print(df,end='\n\n')

df.iloc[ind[0],2]='Fresher'
print(df)
```

```python
print(df.EmpID.isnull())
df.EmpID.isnull().sum()
```

```python
temp = df.dropna().reset_index(drop=True)
temp
```

```python
df.info()
```

```python
# handling duplicate values
print(df,end='\n\n')

ind = np.where(df.Passport.duplicated())
print(ind)

ind = np.where(df.Passport.duplicated(keep=False))  # we have to use keep=False because without this we only get 2nd occurance.  mostly used
print(ind)

ind = np.where(df.Passport.duplicated(keep='first'))  # first occurance reported
print(ind)

ind = np.where(df.Passport.duplicated(keep='last'))   # last occurance is not reported
print(ind)
```
