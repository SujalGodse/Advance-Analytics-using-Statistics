```python
import pandas as pd
import numpy as np
import sklearn
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_excel("/content/drive/MyDrive/CDAC/CDAC_DataBook.xlsx",sheet_name="diabetes")
df.columns
```

```python
df = df[['Glucose','BloodPressure','Age','DietType','Outcome']]
diet_dummy = pd.get_dummies(df.DietType, drop_first=True).astype(int)
df = df.drop('DietType', axis=1)
df = pd.concat([df,diet_dummy], axis=1)
```

```python
df.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(df.drop('Outcome', axis=1), df.Outcome, test_size=0.2, random_state=20)
```

```python
import tensorflow as tf
from tensorflow import keras

from tensorflow.keras.models import Sequential # helps with the forward and backward propagation
from tensorflow.keras.layers import Dense # this is for the hidden, input and output layers
from tensorflow.keras.layers import ReLU   # activation function
from tensorflow.keras.layers import Dropout   # switching off the neurons
```

```python
classifier = Sequential()
classifier.add(Dense(units=5, activation='relu', input_dim=5))    # input layer of 5 neurons
classifier.add(Dense(units=7, activation='relu'))                 # first hidden layer with 7 neurons
classifier.add(Dense(units=5, activation='relu'))                 # second hidden layer with 5 neurons
classifier.add(Dense(units=1, activation='sigmoid'))              # output layer
```

```python
classifier.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
mod1 = classifier.fit(x_train, y_train, batch_size=10, epochs=10)
```

```python
y_pred = classifier.predict(x_test)
y_pred = (y_pred >= 0.5)

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```
