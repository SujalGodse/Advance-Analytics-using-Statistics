```python
df = pd.read_excel("/content/drive/MyDrive/CDAC/CDAC_DataBook.xlsx",sheet_name='Health')
df
```

```python
# broad long - no. of rows gets increased   -> melt()
# broad wide - no. of columns gets increased    -> pivot_table()

df1 = df.melt(id_vars=['Country','Disease'],var_name='Year',value_name='Cases')
df1.head()

# the columns which we didnt want to mofify or update are in id_vars=[]
# the columns which we have to be update are mentioned in var_name='' and value_name=''.
```

```python
print(df.shape)
print(df1.shape)    # Here broad long takes place
```

```python
df2 = df1.pivot_table(values='Cases',index=['Country','Year'],columns='Disease')  # here we have multiindexes
df2
```

```python
df3 = df1.pivot_table(values='Cases',index='Country',columns=['Disease','Year'])
df3
```

```python
print(df1.shape)
print(df2.shape)
print(df3.shape)
```

```python
print(df2.index)
print(df2.columns)
```

```python
df2 = df1.pivot_table(values='Cases',index=['Country','Year'],columns='Disease').reset_index()    # removed multiindex
df2
```

```python
df2 = df1.pivot_table(values='Cases',index=['Country','Year'],columns='Disease').reset_index().rename_axis(columns=None)    # removed index column name
df2
```

### **Statistics**

## **Types of data**
**Qualitative / Categorical / Attribute data:** will be mostly in words (ex: name,state)   
**Quantitative data:** will be in numbers (ex: age,marks)

**Quantitative Parts :**                                                           
**Continuous data ->** bold text Continuous data is the one where possibility of decimal  point cannot be ruled out
*   it will always a result of measurement
*   it will  always be expressed as mlt

**Discrete data ->** will never have decimal point
*   it will always be a result of counting and will never have decimal point


**when n*p and n*(1-p) > 10 then discrete data can be considered as Continuous. (n = sample size)**
**Any rating or ranking number in categorical data**

## **Types of Scale**
1. Nominal : answer will not change if we change logical order . Categories with no logical order
2. Ordinal : answer will change if we change logical order. Categories have logical order - ranting/ranking
3. Interval : 0 does not mean absence of measurement so mathematical operations will be restricted. (ex: temperature). We can do comparision but we cannot take ratios in interval scale.
4. Ratio : used for both count and discrete

-- When type of data is categorical then type of scale will be either Nominal or Ordinal
--


ration : division of two continuous values

proportion : division of two discrete values
