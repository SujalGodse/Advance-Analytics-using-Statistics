```python
import numpy as np
import pandas as pd
from numpy import random
```

```python
# grouping will never be done on continuous data or discrete, it will be done on categorical data

grp1 = df.groupby('MaterialID')
grp1.groups
```

```python
grp1.get_group('SDRT-67')
```

```python
grp1.Quantity.mean()
```

```python
grp1.Quantity.agg(np.mean)
```

```python
grp1.Quantity.agg(np.size)
```

```python
grp1.Quantity.agg([np.size,np.sum,np.mean])
```

```python
grp1.Quantity.agg(['size','sum','mean'])
```

```python
s1 = grp1.Quantity.agg(np.size)
d1 = pd.DataFrame(s1)
d1
```
