```python
df = pd.read_excel('/content/drive/MyDrive/CDAC/CDAC_DataBook.xlsx',sheet_name='PCA_LDA_Data')

x = df.drop(columns='y')
y = df.y

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=20)
```

```python
sc1 = StandardScaler()
x_train_sc = sc1.fit_transform(x_train)
x_test_sc = sc1.transform(x_test)
```

```python
x_train_sc = pd.DataFrame(x_train_sc,columns=x_train.columns,index=x_train.index)
x_test_sc = pd.DataFrame(x_test_sc,columns=x_test.columns,index=x_test.index)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
x_train_pca = pca.fit_transform(x_train_sc)
x_test_pca = pca.transform(x_test_sc)
```

```python
x_train_pca = pd.DataFrame(x_train_pca,columns=['PC1','PC2'],index=x_train.index) # follow original index
x_test_pca = pd.DataFrame(x_test_pca,columns=['PC1','PC2'],index=x_test.index)
```

```python
x_train_pca = sm.add_constant(x_train_pca,prepend=False)
x_test_pca = sm.add_constant(x_test_pca,prepend=False)
```

```python
mod2 = sm.Logit(y_train,x_train_pca).fit()
print(mod2.summary())
```

```python
y_pred = mod2.predict(x_test_pca)
y_pred = (y_pred>=0.5).astype(int)
print(confusion_matrix(y_test,y_pred))
```

# **LDA (Linear Discriminant Analysis)**

```python
df = pd.read_excel('/content/drive/MyDrive/CDAC/CDAC_DataBook.xlsx',sheet_name='PCA_LDA_Data')

preds = df.drop('y', axis=1)
resp = df.y

x_train, x_test, y_train, y_test = train_test_split(preds,resp,test_size=0.2, random_state=20)

sc1 = StandardScaler()

x_train_sc = sc1.fit_transform(x_train)
x_test_sc = sc1.transform(x_test)

x_train_sc = pd.DataFrame(x_train_sc, columns=x_train.columns, index=x_train.index)
x_test_sc = pd.DataFrame(x_test_sc, columns=x_test.columns, index=x_test.index)
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
lda = LinearDiscriminantAnalysis()


x_train_lda = lda.fit_transform(x_train_sc,y_train)
x_test_lda = lda.transform(x_test_sc)

x_train_lda = pd.DataFrame(x_train_lda, columns=['LD1'], index=x_train.index)
x_test_lda = pd.DataFrame(x_test_lda, columns=['LD1'], index=x_test.index)

x_train_lda = sm.add_constant(x_train_lda, prepend=False)
x_test_lda = sm.add_constant(x_test_lda, prepend=False)

mod3 = sm.Logit(y_train, x_train_lda).fit()
x_test_lda = sm.add_constant(x_test_lda, prepend=False)
y_pred = mod3.predict(x_test_lda)
y_pred = (y_pred>=0.5).astype(int)
print(confusion_matrix(y_test,y_pred))
```
