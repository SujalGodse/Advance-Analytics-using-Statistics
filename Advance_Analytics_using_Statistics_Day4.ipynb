Merging DataFrame

```python
mumbai = pd.read_excel("CDAC_DataBook.xlsx",sheet_name='Mumbai')
pune = pd.read_excel("CDAC_DataBook.xlsx",sheet_name='Pune')
```

```python
mumbai
```

```python
pune
```

```python
pd.merge(pune,mumbai)   # it will show empty df because there is no common row between two dfs
```

```python
pd.merge(pune,mumbai,on='Subject')
```

```python
pd.merge(pune,mumbai,on=['Subject','Grade'])
```

```python
# inner join
pd.merge(pune,mumbai,on='Subject',how='inner')
```

```python
# outer join
pd.merge(pune,mumbai,on='Subject',how='outer')
```

```python
# left join
pd.merge(pune,mumbai,on='Subject',how='left')
```

```python
# right join
pd.merge(pune,mumbai,on='Subject',how='right')
```

```python
pd.merge(pune,mumbai,on=['Subject','Grade'],how='outer')
```
