### **Apply PCA**

```python
df = pd.read_excel('/content/drive/MyDrive/CDAC/pca.xlsx')
df.head()

x = df.drop(columns=["Response"])
y = df.Response
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=20)
```

**Standardizing values**

```python
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
```

**Correlation Matrix**

```python
cov_matrix = np.cov(x_train_scaled.T)
print(cov_matrix)
```

**Calculating EigenValues and EigenVectors using covariance methos**

```python
eigen_values,eigen_vectors = np.linalg.eig(cov_matrix)
print(eigen_values)
print()
print(eigen_vectors)
```

```python
df = pd.read_excel('/content/drive/MyDrive/CDAC/pca.xlsx')
df.head()
x = df.drop(columns=["Response"])
y = df.Response
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=20)
```

```python
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
pca = PCA(n_components=2)
xtrain_pca = pca.fit_transform(x_train_scaled)
xtest_pca = pca.transform(x_test_scaled)
```

```python
xtrain_pca = pd.DataFrame(xtrain_pca,columns=['PC1','PC2'],index=x_train.index)
xtest_pca = pd.DataFrame(xtest_pca,columns=['PC1','PC2'],index=x_test.index)
```

```python
xtrain_pca.head()
```

```python
x_train_pc1 = sm.add_constant(xtrain_pca, prepend=False)
x_test_pc1 = sm.add_constant(xtest_pca, prepend=False)
```

```python
from statsmodels.discrete.discrete_model import MNLogit
mod2pca = sm.Logit(y_train,x_train_pc1).fit()
```

```python
y_pred = mod2pca.predict(x_test_pc1)
y_pred = (y_pred>=0.5).astype(int)
print(confusion_matrix(y_test,y_pred))
```
