```python
n = 300
xbar = 1800
std = 70
cl = 0.95 # 95%
```

```python
norm.ppf((1-cl)/2)
```

```python
z = np.abs(norm.ppf((1-cl)/2))
z
```

```python
t = np.abs(t.ppf((1-cl)/2,n-1))
t
```

```python
# use t because it will work on both small and big sample size 30. z is used for greater then 30
low_est = xbar - t*std/n**0.5       # XA
low_est
```

```python
upper_est = xbar + t*std/n**0.5     # XB
upper_est
```

```python
size = upper_est - low_est
size
```

```python
moe = (upper_est - low_est)/2    # Margin of Error
moe
```

```python
moe1 = (t*std)/n**0.5   # Margin of Error
moe1
```

```python
n1 = ((t**2) * (std**2)) / moe**2   # n == n1 sample size
n1
```
