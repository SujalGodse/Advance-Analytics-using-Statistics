# **Dispersion or variation or spread -**
range , inter quartile range, variance

Range : max-min                                                                   
Problems with range:
1. affected by outliers
2. it uses only 2 values from list and does not care about other values

inter quartile range : 75th  percentile - 25th percentile (ex: if 75 percentile of data is assume 40 then 75% of them are below marks)                                                  
Q3 = 75th percentile                                                            
Q2 = median

Why IQR better then range ?                             
IQR fills the range because it removes the outlier

formula of percentile = i = 1 + (n-1)*p/100

```python
IQR = np.percentile(x1,75) - np.percentile(x1,25)
IQR
```

# **Probability Distribution**

```python
from scipy.stats import binom
```

Chances of rain on any given day in june in mumbai is 0.7

a. it rains for exactly 10 days
b. it rains for maximum 10 days
c. it rains for minimum 10 days

```python
binom.pmf(10,15,0.7)      # exactly 10 days
```

```python
binom.cdf(10,15,0.7)      # maximum 10 days
```

```python
#  c. it rains for minimum 10 days
# in this case days included will be 11,12,13,14,15 + 10 (10 will also be included) i.e 10,11,12,13,14,15
print(1 - binom.cdf(10,15,0.7)) # it will answer for 11,12,13,14,15
print(1 - binom.cdf(9,15,0.7)) # it will answer for 10,11,12,13,14,15  # result of c
```

```python
# Poisson Distribution
# Q. avg number of occurences is 15. what is the probability of
from scipy.stats import poisson
```

```python
# a. exactly 12 incidents taking place today
poisson.pmf(12,15)
```

```python
# 3 b. upto 12 incidents take place
poisson.cdf(12,15)
```

```python
# c. minimum 12 incidents take place
1 - poisson.cdf(11,15)
```

```python
poisson.sf(11,15)     # its same like above one
```

```python
# Normal Distribution - Bell shaped Graph
# mean = 16hrs , std = 2.3hrs
# Q1. <=14hrs, Q2. >=14hrs Q3. 13<= <=14
from scipy.stats import norm,t
# Q1
print(norm.cdf(14,16,2.3))
# Q2
print(1 - norm.cdf(14,16,2.3))
print(norm.sf(14,16,2.3))
# Q3
norm.cdf(17,16,2.3) - norm.cdf(13,16,2.3)
```
