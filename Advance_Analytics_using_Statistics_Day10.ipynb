## 1-sample z test/ 1 sample t test

Used to check whether the difference betweent the sample mean and claim value can be ignored or not.

Used to check whether the difference betweent the sample mean and claim value is statistically significant or not

step 1:

1. H0 U >= 3.7
2. Ha U < 3.7

Step 2:

CL = 95% = 0.95
LoS = 5% = 0.05

Step 3: 1 sample z test

Step 4: TS,p-value

```python
# finding TS and p-value
t = ssw.ztest(erupt,value=3.7,alternative='smaller')    # returns (TS,p-value)
t
```

```python
# compare p-value and LoS=0.05
if t[1] > 0.05:
  print("Do not Reject H0")
else:
  print("Reject H0")
```
