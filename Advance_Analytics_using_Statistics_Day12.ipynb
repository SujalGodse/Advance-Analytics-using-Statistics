# **2-proportion test**

**Q. Pune CDAC claims that they have atleast 8% more  engineers as compared to CDAC kharghar**

H0 pr(Pune) - pr(Kharghar) >= 0.08

Ha pr(Pune) - pr(Kharghar) < 0.08

```python
# In a sample of 70 students in pune, 40 are Btechs
# In a sample of 80 students in Kharghar, 48 were Btechs
40/70 - 48/80
```

```python
engg = np.array([40,48])
total = np.array([70,80])
statsmodels.stats.proportion.proportions_ztest(engg,total,0.08,alternative='smaller')
```

# **Chi-Square test for goodness of fit**

Q. claimed pattern for the qualifiication

Btech - 0.4, Mtech - 0.1, MCA - 0.3, Msc - 0.2

Btech - 50, Mtech - 20, MCA - 45, Msc - 25

whether counts obtained from sampole follow the pattern or not

```python
from scipy.stats import chisquare
```

```python
counts = np.array([50,20,45,25])
pattern = np.array([0.4,0.1,0.3,0.2])
```

```python
# if pattern was followed what is expected counnt
exp_counts = sum(counts) * pattern
exp_counts
```

```python
scipy.stats.chisquare(counts,f_exp=exp_counts)
# H0 do not rejected
```

```python
counts = np.array([70,10,35,45])
pattern = np.array([0.4,0.1,0.3,0.2])
exp_counts = sum(counts) * pattern
print(exp_counts)
scipy.stats.chisquare(counts,f_exp=exp_counts)
# H0 rejected
```

```python
sum((exp_counts-counts)**2/exp_counts)    # test statistics
```

```python
(exp_counts-counts)**2/exp_counts
```
