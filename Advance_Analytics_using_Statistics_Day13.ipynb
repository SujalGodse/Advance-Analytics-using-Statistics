# **One Sample Poisson Test**

Q. a good class will ask atleast 20 question everyday

H0 n >= 20

Ha n < 20

Left tail test

```python
from statsmodels.stats import rates
from statsmodels.stats.rates import test_poisson
```

```python
questions = [14,19,24,18,17,23,21,18,19,17]
np.mean(questions)
```

```python
test_poisson(sum(questions),10,20,method='score',alternative='smaller')
```

# **One Sample Poisson Rate**

Q. DbDA batch asks atleast 6 questions/day more than AI batch


H0 : avg_count(DBDA) - avg_count(AI) >= 6

Ha : avg_count(DBDA) - avg_count(AI) < 6

In DBDA's 12 days of training, 145 questions were asked

In AI's 5 days of training , 45 questions were asked

```python
from statsmodels.stats.rates import test_poisson_2indep
test_poisson_2indep(145,12,45,5,value=6,compare='diff',method='score',alternative='smaller')
```

# **One Variance Test**

```python
df = pd.read_excel('/content/drive/MyDrive/CDAC/CDAC_DataBook.xlsx',sheet_name='faithful')
erupt = df.eruptions
```

```python
np.var(erupt,ddof=1)
```

H0 variance(erupt) < 1.1

H0 variance(erupt) >= 1.1

Ha variance(erupt) > 1.1  --> right tail test

Test_Statistics = (n-1)*variance/(claimed value)

```python
ts = (len(erupt)-1)*np.var(erupt,ddof=1)/1.1    # calculating test statistics
ts
```

```python
from scipy.stats import chi2
p_value = 1 - chi2.cdf(ts,len(erupt)-1)   # getting p-value from test statistic
p_value
```

```python
p_value = chi2.sf(ts,len(erupt)-1)
p_value
```

# **Two Variance Test**

H0: var(samp1)/var(samp2) >= 1.4

H0: var(samp1)/var(samp2) < 1.4


```python
samp1 = erupt[30:70]
samp2 = erupt[130:180]
np.var(samp1,ddof=1)/np.var(samp2,ddof=1)
```

```python
samp_ratio = np.var(samp1,ddof=1)/np.var(samp2,ddof=1)
ts = samp_ratio/1.4
```

```python
from scipy.stats import f
# f test is used whenever we are taking ratio of two variances
f.cdf(ts,len(samp1)-1,len(samp2)-1)
```
